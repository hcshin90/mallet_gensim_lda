```python
#from gensim.test.utils import common_corpus, common_dictionary
from gensim.models.wrappers import LdaMallet
import gensim
from gensim import corpora
import logging
import ldamallet_params
import pickle as pk
import pyLDAvis.gensim as gensimvis
import pyLDAvis
```

```python
print('Loading Corpus, Dictionary Pickle..')
with open('./output/corpus_.pickle', 'rb') as f:
    corpus = pk.load(f)

with open('./output/dictionary_.pickle', 'rb') as f:
    dictionary = pk.load(f)
```

```python
print('Loading Gensim_LDA Model..')
loaded_model = gensim.models.ldamodel.LdaModel.load('./output/gensim_model')
```

```python
print('Making LDAvis using Gensim_LDA_Model, Corpus, Dictionary..')
vis_data = gensimvis.prepare(loaded_model, corpus, dictionary)
```
